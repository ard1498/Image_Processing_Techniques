```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
im_name1 = input('enter the image 1 :')
image1 = cv2.imread(im_name1)

cv2.imshow('Original Image 1', image1)
cv2.waitKey(0)
cv2.destroyAllWindows()

im_name2 = input('enter the image 2 :')
image2 = cv2.imread(im_name2)

cv2.imshow('Original Image 2', image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def getAddn(image1, image2):
    new_img = np.zeros(image1.shape)
    n = image1.shape[0]
    m = image2.shape[1]
    for i in range(n):
        for j in range(m):
            new_img[i,j][0] = (image1[i,j][0] + image2[i,j][0])
            new_img[i,j][1] = (image1[i,j][1] + image2[i,j][1])
            new_img[i,j][2] = (image1[i,j][2] + image2[i,j][2])
    return new_img
```

```python
# addn_image = getAddn(image1.copy(), image2.copy())
# addn_image = cv2.add(image1.copy(), image2.copy())
addn_image = image1.copy() + image2.copy()
```

```python
cv2.imshow('added image',addn_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
_, ax = plt.subplots(nrows = 3, ncols= 3, figsize=(60,60))
ax[0,0].imshow(addn_image)
ax[0,1].imshow(image1)
ax[0,2].imshow(image2)
plt.show()
```
